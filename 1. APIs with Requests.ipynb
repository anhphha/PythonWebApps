# Another example: the iTunes Search API

## Passing parameters in the request

```python
base_site = "https://itunes.apple.com/search"
```

```python
url = base_site+"?term=the+beatles&country=fi"
requests.get(url)
```

```python
r = requests.get(base_site, params = {'term':'the beatles','country':'us'})
r.url
```

### Investigating the output and parameters

```python
info = r.json()
print(json.dumps(info,indent = 3))
```

```python
info.keys()
```

```python
print(json.dumps(info['results'][-1], indent=4))
```

```python
info['resultCount']
```

```python
r = requests.get(base_site, params = {'term':'the beatles','country':'fi','limit':200})
r.ok
```

```python
r.json()['resultCount']
```

```python
r = requests.get(base_site,params={'media':'hahaha','term':'alternative','limit':200})
r.ok
```

```python
r.json()#['error']
```

```python
r.json()['errorMessage']
```

### Structuring and Exporting the Data

```python
r = requests.get(base_site,params={'term':'jack johnson','entity':'musicVideo'})
r.url
```

```python
info = r.json()
print(json.dumps(info,indent = 3))
```

```python
info['results'][0]['trackName']
```

```python
info['results'][0]['releaseDate']
```

```python
for i in range(len(info['results'])):
    print('Track Name:','of track',i,info['results'][i]['trackName']),'\n'
    print('Release Date:','of track',i,info['results'][i]['releaseDate'])
    print('\n')
```

```python
### Using pandas
import pandas as pd
songs_df = pd.DataFrame(info['results'])
```

```python
songs_df
```

```python
songs_df.to_csv('songs_info.csv')
```

```python
songs_df.to_excel('songs_info.xlsx')
```
