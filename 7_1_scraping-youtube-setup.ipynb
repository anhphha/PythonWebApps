# Scraping YouTube

## Initial Setup

```python
from bs4 import BeautifulSoup
import requests
```

## Connect to webpage

```python
url = "https://www.youtube.com/"
```

```python
r = requests.get(url)
r.status_code
```

```python
# get HTML
html = r.text
```

```python
html
```

```python
# convert HTML to BeautifulSoup object
soup = BeautifulSoup(html,'lxml')
```
